```
#load dataset
data_url = 'https://d37ci6vzurychx.cloudfront.net/trip-data/yellow_tripdata_2024-11.parquet'
data = pd.read_parquet(data_url)
data.head()
```

```
#create a spark session
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("NYC Taxi Data Analysis").getOrCreate()
```

```
spark_df = spark.createDataFrame(data)
```

```
#define the shape of the dataset
num_row = spark_df.count()#this will be for the number of rows
#get no of coluumns
num_columns = len(spark_df.columns)
#print shape of the dataset
print(f"Number of rows: {num_row}")
print(f"Number of columns: {num_columns}")
```
